This code displays all the coordinates that were unmatched from the ground truth file and the csv file, containing all the features.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import csv
```

```python
original_file = r'C:/Users/T00701453.MYTRU/Downloads/combined.csv'  # all coordinates
converted_file = r'C:/Users/T00701453.MYTRU/Downloads/SA_ground_labelled.csv'  # ground truth

# Load the CSV files into DataFrames
original_df = pd.read_csv(original_file)
converted_df = pd.read_csv(converted_file)

# Ensure both DataFrames have 'Latitude' and 'Longitude' columns
if 'Latitude' in original_df.columns and 'Longitude' in original_df.columns \
        and 'Latitude' in converted_df.columns and 'Longitude' in converted_df.columns:
    
    # Perform an inner merge to find matching rows in both files based on 'latitude' and 'longitude'
    matches_df = pd.merge(converted_df, original_df, on=['Latitude', 'Longitude'], how='inner')
    
    # Find unmatched coordinates
    unmatched_in_original = original_df.merge(matches_df[['Latitude', 'Longitude']], on=['Latitude', 'Longitude'], how='left', indicator=True)
    unmatched_in_original = unmatched_in_original[unmatched_in_original['_merge'] == 'left_only'][['Latitude', 'Longitude']]
    
    unmatched_in_converted = converted_df.merge(matches_df[['Latitude', 'Longitude']], on=['Latitude', 'Longitude'], how='left', indicator=True)
    unmatched_in_converted = unmatched_in_converted[unmatched_in_converted['_merge'] == 'left_only'][['Latitude', 'Longitude']]

    # Check if there are any matching rows
    if not matches_df.empty:
        print("Matching coordinates found:")
        print(matches_df)
    else:
        print("No matching coordinates found.")

    # Print unmatched coordinates
    if not unmatched_in_original.empty:
        print("Unmatched coordinates in the original file:")
        print(unmatched_in_original)
    else:
        print("No unmatched coordinates in the original file.")

    if not unmatched_in_converted.empty:
        print("Unmatched coordinates in the converted file:")
        print(unmatched_in_converted)
    else:
        print("No unmatched coordinates in the converted file.")

else:
    print("Error: 'Latitude' and 'Longitude' columns not found in both files.")
```
